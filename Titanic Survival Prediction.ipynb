```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('C:/Users/rgangurde/Downloads/Data/trainTitanic.csv')
```

```python
df.head()
```

```python
X= df.drop('PassengerId',axis=1).drop('Survived',axis=1).drop('Name',axis=1).drop('Ticket',axis=1).drop('Fare',axis=1).drop('Cabin',axis=1)
```

```python
Y= df['Survived']
```

```python
from sklearn.preprocessing import LabelEncoder

lb= LabelEncoder().fit(['male','female'])
X['Sex']= lb.transform(df['Sex'])
```

```python
import math
str= 'wow'
avgAge= np.mean(X['Age'])
for i in range(len(X)):
    if type(X['Embarked'][i]) != type(str):
        X['Embarked'][i]= 'N'
    if math.isnan(X['Age'][i]):
        X['Age'][i]= avgAge
```

```python
X
```

```python
lb= LabelEncoder().fit(['S','C','Q','N'])
X['Embarked']= lb.transform(X['Embarked'])
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(X, Y, test_size= 0.5, random_state= 2)
```

```python
from sklearn.tree import DecisionTreeClassifier
dTree= DecisionTreeClassifier(
    criterion='gini', 
    splitter= 'best',
    max_depth= 5,
    max_features= 4,
    min_impurity_split= 0.2,
    min_samples_split= 2)

dTree.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
yHat= dTree.predict(x_test)
print('Accuracy:', accuracy_score(y_test, yHat))
```
