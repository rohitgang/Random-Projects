### Imports

```python
import numpy as np
from scipy import misc
import imageio
from skimage import data, io, filters
from matplotlib import pyplot as plt
from random import shuffle
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import sqlite3
from training import image_training as it
```

### SQLITE3 Connection info

```python
database = '//itdhwy/DESFiles/DES/TransportationSystems/Common/ImageRecognition/Database/ImageRecognition.db'
database_connection = sqlite3.connect(database)
cursor = database_connection.cursor()
```

### Function to grab classified data

```python
def get_flag(cursor):
    filepath_flag_lt = []
    query = (
    """
    SELECT Images.image_filepath, CompletedImages.guardrail_present
    FROM Images INNER JOIN CompletedImages
    ON Images.image_id = CompletedImages.image_id;
    """
    )
    cursor.execute(query)

    for row in cursor:
        filepath_flag_lt.append((row[0], row[1]))

    return filepath_flag_lt
```

### Grab classified data

```python
classified_data = get_flag(cursor)
print (len(classified_data))
```

## test files

```python
classified_true = it.get_true_images('guardrail')
print(len(classified_true))
```

### Add test list and classified lists together

```python
lt = [classified_data[0:300], classified_true[0:300]]

flattened_list = [i for s in lt for i in s]

print (flattened_list)
print(len(flattened_list))
```

## Shuffle test files

```python
shuffle(flattened_list)
for i in flattened_list[0:100]:
    print (i)
    
```

### Turn images and values into a dictionary of arrays

```python
dictionary = {}
answers    = []
data       = []
i = 0
for item in flattened_list:
    try:
        data_array = np.array(imageio.imread(item[0]))
        data.append(data_array)
        answers.append(item[1])
    except:
        print ('Item failed {0}'.format(item[0]))
dictionary['target'] = np.array(answers)
dictionary['data'] = np.array(data)
```

### Image set basics

```python
number_of_images = len(dictionary['data'])
print ('Number of images: {0}'.format(number_of_images))
```

### Extract images and targets

```python
X = dictionary['data']
y = dictionary['target']

print (X.shape)
```

### Random Forest Classifier

```python
clf = RandomForestClassifier()
```

### Reshape the data

```python
X = X.reshape(X.shape[1] * X.shape[2] * X.shape[3], X.shape[0]).T
y = y.reshape(y.shape[0],)


print (X.shape)
print (len(y))
```

### Train the model

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
clf.fit(X_train, y_train)
```

### Test the model based on prediction set

```python
preds = clf.predict(X_test)
print("Accuracy:", accuracy_score(y_test,preds))
print (X_test.shape)
```
